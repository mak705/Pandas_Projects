```python
import pandas as pd
import numpy
```

```python
csv_url = 'C:\\Users\\makbulhussain.k\\Downloads\\student-mat.csv'
df = pd.read_csv(csv_url)
df.head()
```

```python
#For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column
stud_alcoh = df.loc[: , "school":"guardian"]
stud_alcoh.head()
```

```python
captalizer = lambda x: x.upper()
```

```python
stud_alcoh['Mjob'].apply(captalizer)
stud_alcoh['Fjob'].apply(captalizer)
# Print the last elements of the data set.
stud_alcoh.tail()
```

```python
#Create a function called majority that return a boolean value to a new column called legal_drinker 
#(Consider majority as older than 17 years old)
def majority(x):
    if x > 17:
        return True
    return False
```

```python
#Create a New Column legal_drinker and apply majority
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head()
```

```python
#Simply Multiply every number of the dataset by 10
def times10(x):
    if type(x) is int:
        return 1 * x
    return x
```
